```python
import pandas as pd
import matplotlib.pyplot as plt

data_dict = {
    "Date": ["2025-01-01", "2025-01-02", "2025-01-03", "2025-01-04", "2025-01-05"],
    "Product": ["Phone", "Laptop", "Tablet", "Phone", "Laptop"],
    "Region": ["North", "South", "East", "West", "North"],
    "Sales": [1200, 2500, 900, 1500, 2300],
    "Quantity": [3, 2, 4, 3, 1]
}

df = pd.DataFrame(data_dict)
df.to_csv("sample_sales.csv", index=False)
print(" Sample file 'sample_sales.csv' created successfully!\n")

data = pd.read_csv("sample_sales.csv")

print(" Total Sales:", data["Sales"].sum())
print("\n Sales by Product:\n", data.groupby("Product")["Sales"].sum())
print("\n Sales by Region:\n", data.groupby("Region")["Sales"].sum())

data.groupby("Product")["Sales"].sum().plot(
    kind="bar", title="Sales by Product", color="skyblue", edgecolor="black"
)
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.show()
```
